***Using open ai, groq, Gemini***

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["Openai_Api_key"]=os.getenv("openai")
os.environ["Groq_Api"]=os.getenv("groq")
os.environ["Gemini_Ai"]=os.getenv("googleaistudio")
```

```python
from langchain.chat_models import init_chat_model
model=init_chat_model("gpt-4.1")
model
```

```python
r=model.invoke("Hello, How are you?")
print(r.content)

#no billing detail in open ai api so no use currently using open ai api
```

```python
from langchain.chat_models import init_chat_model

os.environ["Google_api_key"]=os.getenv("googleaistudio")

model=init_chat_model("google_genai:gemini-2.5-flash-lite")
r=model.invoke("why parrot talk")
print(r.content)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model=ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")

r=model.invoke("How long will it take for me to learn a 12 hr yt vid")

print(r.content)
```
